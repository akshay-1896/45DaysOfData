<h3>Splitting Dependent and Independent Variables</h3>

```python
x = df2.drop('price',axis='columns') # Independent(Feature) Variables
y = df2[['price']] # Dependent(Target) Variables
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) # test size is 20%
x_train
```

```python
# importing linear regression algorithm class
from sklearn.linear_model import LinearRegression
```

```python
# instance creation
lnr = LinearRegression()
```

```python
# algorithm training
lnr.fit(x_train,y_train)
```

```python
# Model evaluation
# lnr.score(x_train,y_train)*100 # Should be high(90 ~ 95%) for better predictions on data
print(f"Training score: {lnr.score(x_train,y_train)*100}")
print(f"Testing score: {lnr.score(x_test,y_test)*100}")
```

<p>From above scores, it could be seen that difference between training and testing score is less than 5%

Therefore, it could be concluded that this is a <span style:font-weight="bolder">Generalised Model</span></p>

```python
x_test.shape
# total 1462 bike data
```

```python
# Predicton or testing phase 
prediction = lnr.predict(x_test)
```

```python
prediction.shape
```

```python
compare_df = pd.concat([x_test,y_test],axis=1)
```

```python
compare_df['prediction'] = prediction
```

```python
compare_df.head(40)
```

```python
compare_df['residual'] = abs(compare_df['price'] - compare_df['prediction'])
compare_df.head()
```

```python
# Min residual means best prediction

min_row = compare_df.loc[compare_df['residual'].idxmin()]
min_row

# compare_df['residual'].idxmin() # gives row index with min value of selected column

# compare_df.loc[] # gives entire row data with respect to entered index 
```
