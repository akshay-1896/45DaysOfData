```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

print("Libraries imported!!")
```

```python
country_population = [700,750,740,730,752,800,790]
plt.plot(country_population)
plt.show()
```

```python
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.plot(x_axis,country_population,marker=".",ms = 15,mfc = 'red', color = 'blue', ls='dashdot', lw = 2) 
# ms = marker size, mfc = marker face color
# ls = line style, lw = line width
plt.title('Line Graph Plot')
plt.xlabel('Axis-X')
plt.ylabel('Country Population')
plt.show()
```

<h1>Bar Plot</h1>

```python
country_population = [450,730,900,450,800]
country_name = ['Japan', 'Bangladesh', 'India', 'Nepal', 'China']

plt.bar(country_name,country_population, color = 'cyan', edgecolor = 'blue', linewidth = 2, label= 'Country Population')
plt.legend(bbox_to_anchor=(1.15,1.15))
plt.xlabel('Country Name')
plt.ylabel('Country Population')
plt.show()
```

```python
df = pd.read_csv('Used_Bikes.csv')
```

```python
df.drop_duplicates(inplace=True)
df['brand'].value_counts()
```

```python
brand_name = list(df['brand'].value_counts().keys())
brand_name
```

```python
no_of_count = list(df['brand'].value_counts().values)
no_of_count
```

```python
plt.figure(figsize=(10,4)) # (width , height)
brand_name = list(df['brand'].value_counts().keys())
no_of_count = list(df['brand'].value_counts().values)
plt.bar(brand_name,no_of_count, hatch = '///', alpha= 0.5, edgecolor= 'blue', label= 'Used Bikes')
plt.xticks(rotation=80)
# plt.yticks(rotation=)
plt.xlabel('Bike Name')
plt.ylabel('Availability')
plt.legend()
plt.savefig('brand_population', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,4))
df['owner'].value_counts()
owner = list(df['owner'].value_counts().keys())
no_of_owner = list(df['owner'].value_counts().values)
plt.bar(owner, no_of_owner, width=0.4, label= 'Owners')
plt.legend()
plt.show()
```
